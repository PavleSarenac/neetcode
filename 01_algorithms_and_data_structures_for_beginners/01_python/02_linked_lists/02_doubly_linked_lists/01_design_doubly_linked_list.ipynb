```python
class ListNode:
    def __init__(self, value, previous = None, next = None):
        self.value = value
        self.previous = previous
        self.next = next

class MyLinkedList:

    def __init__(self):
        dummyNode = ListNode(-1)
        self.head = self.tail = dummyNode

    def get(self, index: int) -> int:
        currentNode = self.head.next
        currentIndex = 0
        while currentNode:
            if currentIndex == index:
                return currentNode.value
            currentNode = currentNode.next
            currentIndex += 1
        return -1
            

    def addAtHead(self, val: int) -> None:
        newNode = ListNode(val, self.head, self.head.next)
        if self.head.next:
            self.head.next.previous = newNode
        else:
            self.tail = newNode
        self.head.next = newNode

    def addAtTail(self, val: int) -> None:
        newNode = ListNode(val, self.tail)
        self.tail.next = newNode
        self.tail = newNode

    def addAtIndex(self, index: int, val: int) -> None:
        currentNode = self.head.next
        currentIndex = 0
        while currentNode:
            if index == currentIndex:
                newNode = ListNode(val, currentNode.previous, currentNode)
                currentNode.previous.next = newNode
                currentNode.previous = newNode
                return
            currentNode = currentNode.next
            currentIndex += 1
        if index == currentIndex:
            self.addAtTail(val)

    def deleteAtIndex(self, index: int) -> None:
        currentNode = self.head.next
        currentIndex = 0
        while currentNode:
            if index == currentIndex:
                currentNode.previous.next = currentNode.next
                if currentNode.next:
                    currentNode.next.previous = currentNode.previous
                else:
                    self.tail = currentNode.previous
                return
            currentNode = currentNode.next
            currentIndex += 1
```
