```python
class ListNode:
    def __init__(self, value, next = None, previous = None):
        self.value = value
        self.next = next
        self.previous = previous

class Solution:
    def __init__(self):
        self.head = ListNode(-1)
        self.tail = ListNode(-1)
        self.head.next = self.tail
        self.tail.previous = self.head
        self.queueSize = 0

    def createQueue(self, studentPreferences: list[int]):
        for studentPreference in studentPreferences:
            self.pushToQueueEnd(ListNode(studentPreference))
        self.queueSize = len(studentPreferences)

    def pushToQueueEnd(self, node: ListNode):
        node.next = self.tail
        node.previous = self.tail.previous
        self.tail.previous.next = node
        self.tail.previous = node

        self.queueSize += 1

    def removeFromQueue(self, node: ListNode):
        node.previous.next = node.next
        node.next.previous = node.previous

        self.queueSize -= 1

    def countStudents(self, students: list[int], sandwiches: list[int]) -> int:
        self.createQueue(students)
        currentSandwich = 0
        while currentSandwich < len(sandwiches):
            previousSandwich = currentSandwich
            currentStudent = self.head.next
            currentQueueSize = self.queueSize
            for _ in range(currentQueueSize):
                nextStudent = currentStudent.next
                self.removeFromQueue(currentStudent)
                if currentStudent.value == sandwiches[currentSandwich]:
                    currentSandwich += 1
                else:
                    self.pushToQueueEnd(currentStudent)
                currentStudent = nextStudent

            if previousSandwich == currentSandwich:
                break
        return self.queueSize
```

```python
from collections import deque
from typing import List


class Solution:
    def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
        n = len(students)
        q = deque(students)

        res = n
        for sandwich in sandwiches:
            cnt = 0
            while cnt < n and q[0] != sandwich:
                cur = q.popleft()
                q.append(cur)
                cnt += 1

            if q[0] == sandwich:
                q.popleft()
                res -= 1
            else:
                break
        return res
```

```python
class Solution:
    def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
        n = len(students)
        idx = 0

        res = n
        for sandwich in sandwiches:
            cnt = 0
            while cnt < n and students[idx] != sandwich:
                idx += 1
                idx %= n
                cnt += 1

            if students[idx] == sandwich:
                students[idx] = -1
                res -= 1
            else:
                break
        return res
```

```python
from collections import Counter


class Solution:
    def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
        res = len(students)
        cnt = Counter(students)

        for s in sandwiches:
            if cnt[s] > 0:
                res -= 1
                cnt[s] -= 1
            else:
                break

        return res
```
