```python
class Solution:
    def removeElement(self, nums: list[int], val: int) -> int:
        currentWriteIndex = 0
        for currentReadIndex in range(len(nums)):
            if nums[currentReadIndex] != val:
                nums[currentWriteIndex] = nums[currentReadIndex]
                currentWriteIndex += 1
        return currentWriteIndex
```

```python
class Solution:
    def removeElement(self, nums: list[int], val: int) -> int:
        tmp = []
        for num in nums:
            if num == val:
                continue
            tmp.append(num)
        for i in range(len(tmp)):
            nums[i] = tmp[i]
        return len(tmp)
```

```python
class Solution:
    def removeElement(self, nums: list[int], val: int) -> int:
        i = 0
        n = len(nums)
        while i < n:
            if nums[i] == val:
                n -= 1
                nums[i] = nums[n]
            else:
                i += 1
        return n
```
