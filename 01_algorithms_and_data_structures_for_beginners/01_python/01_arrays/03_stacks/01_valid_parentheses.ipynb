```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        for currentCharacter in s:
            if currentCharacter == ')':
                if len(stack) == 0 or stack.pop() != '(':
                    return False
            elif currentCharacter == '}':
                if len(stack) == 0 or stack.pop() != '{':
                    return False
            elif currentCharacter == ']':
                if len(stack) == 0 or stack.pop() != '[':
                    return False
            else:
                stack.append(currentCharacter)
        return len(stack) == 0
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        while '()' in s or '{}' in s or '[]' in s:
            s = s.replace('()', '')
            s = s.replace('{}', '')
            s = s.replace('[]', '')
        return s == ''
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        closeToOpen = { ")" : "(", "]" : "[", "}" : "{" }

        for c in s:
            if c in closeToOpen:
                if stack and stack[-1] == closeToOpen[c]:
                    stack.pop()
                else:
                    return False
            else:
                stack.append(c)

        return True if not stack else False
```
