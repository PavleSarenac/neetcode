```python
class Solution:
    def hasDuplicate(self, nums: list[int]) -> bool:
        scannedNumbers = set()
        for num in nums:
            if num in scannedNumbers:
                return True
            scannedNumbers.add(num)
        return False
```

```python
class Solution:
    def hasDuplicate(self, nums: list[int]) -> bool:
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] == nums[j]:
                    return True
        return False
```

```python
class Solution:
    def hasDuplicate(self, nums: list[int]) -> bool:
        nums.sort()
        for i in range(1, len(nums)):
            if nums[i] == nums[i - 1]:
                return True
        return False
```
